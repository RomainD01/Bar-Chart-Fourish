### Web Scraping en action !!!

````
# Le code suivant nous permet de pointer sur le site web sur lequel nous désirons extraire/collecter les données (countryflags.com)
# Exécutez le code.

url = "https://www.countryflags.com/"
response = get(url)

html_soup = BeautifulSoup(response.text, 'html.parser')
````

````
# Le code suivant nous permet d'extraire/collecter les données depuis le site countryflags.com.
# Exécutez le code.

containers = html_soup.find_all('div', class_='thumb')
for container in containers: 
  try:
    country_name = container.find('span', class_='title').text[8:]
    country_flag_url = container.find('img')['src']
    print(country_name, end=" ")
    print(country_flag_url)
    sleep(0.05)
  except:
    pass
````

````
# Réutilisez le code de la cellule ci-dessus et stockez le nom des pays dans une liste appelée 'countries_name' et les urls des drapeaux dans une liste appelée 'countries_flag'

containers = html_soup.find_all('div', class_='thumb')
countries_name = []
countries_flag = []
for container in containers: 
  try:
    country_name = container.find('span', class_='title').text[8:]
    country_flag_url = container.find('img')['src']
    countries_name.append(country_name)
    countries_flag.append(country_flag_url)
    #print(country_name, end=" ")
    #print(country_flag_url)
    #sleep(0.05)
  except:
    pass
print(countries_name)
print(countries_flag)
````

````
# Créez un dataframe 'df_countries_flag' à partir de nos listes 'countries_name' et 'countries_flag'.

#df_countries_flag = pd.DataFrame({"countries" : countries_name, "flags" : countries_flag})
df_countries_flag = pd.DataFrame(list(zip(countries_name, countries_flag)), columns = ["country_name", "country_flag"])
df_countries_flag.head()
````

### Feature Engineering

````
# Le code suivant vérifie si le nom d'un pays est présent à la fois dans le dataset 'pop_df' et 'df_countries_flag'. Si c'est le cas, alors l'url
# du drapeau du pays est transféré dans le dataset 'pop_df', dans une nouvelle colonne appelée 'url_flag'. Dans le cas contraire, la valeur -1 sera
# ajoutée. Le code applique cela à tous les pays du dataset 'pop_df'.

pop_df['url_flag'] = (pop_df.country.apply(lambda x: df_countries_flag.query('@x in country_name').index.values)
                         .apply(lambda x: x[0] if len(x) else -1)).apply(lambda x: df_countries_flag.country_flag[x] if x != -1 else x)
````

````
# Extraire la colonne 'country' depuis notre dataframe et la transformer en liste
liste_countries_name = list(pop_df["country"])
````

````
# Utilisez la librairie pycountry_convert afin d'associer chaque pays à son code continent. 
# Indice: nous utiliserons ici le format 'alpha2' pour obtenir de meilleurs résultats.
# Afin d'éviter tout arrêt d'exécution de notre code en cas de "non match", utilisez les instructions (statements) 'try except'.
liste_countries_ISO = []
for country in liste_countries_name:
  try:
    liste_countries_ISO.append(pc.country_name_to_country_alpha2(country))
  except:
    liste_countries_ISO.append("NaN")

liste_continents_ISO = []
for ISO in liste_countries_ISO:
  try:
    liste_continents_ISO.append(pc.country_alpha2_to_continent_code(ISO))
  except:
    liste_continents_ISO.append("NaN")
````

````
# Créez un nouveau dataframe qui contiendra les noms des pays et codes continents
# Pour cela, la fonction zip() pourra vous être utile. Documentez-vous pour comprendre comment l'utiliser.

pop_df_2 = pd.DataFrame(list(zip(liste_countries_name, liste_continents_ISO)), columns = ["country_name", "continent_code"])
pop_df_2.head
````

````
# Trouver une méthode avec pandas afin d'extraire la liste des continents UNIQUES de notre nouveau dataframe.
# Il y a t-il des valeurs non renseignées ('NaN')?? Si oui, notre prochaine mission sera de les transformer.

pop_df_2["continent_code"].unique()
````

````
# Faire un filtre (masque) afin de trouver toutes les lignes de notre dataframe contenant des NaN dans la colonne des codes continents.
mask = pop_df_2["continent_code"] == "NaN"
a = pop_df_2[mask]
a
````

````
# Il y a peu de 'NaN'. Nous allons donc pouvoir les remplacer manuellement dans notre liste 'continent_code'.
# Pour cela, recupérez le numéro de ligne (index) de notre dataframe pour les codes continents ayant pour valeur 'NaN'.
# À partir de ces index, remplacer les 'NaN' de notre liste 'continent_code' par les éléments de la liste ci-dessous:

missing_continent_code = pop_df_2[pop_df_2["continent_code"] == "NaN"].index.tolist()
continents_to_add = ['AF', 'AF', 'AF', 'EU', 'AS', 'OC', 'NA', 'AS']

for index in range(len(missing_continent_code)):
  pop_df_2.iloc[missing_continent_code[index]]["continent_code"] = continents_to_add[index]

pop_df_2.head()
````

````
# Afin de limiter le nombre le labels dans notre visualisation, transformez les codes continents en 4 régions du monde 
# ('Asia', 'Europe', 'Africa', 'Americas'). Pour cela, comprenez et exécutez le code suivant:

pop_df.loc[pop_df['continent_code'] == 'NA', 'continent_code'] = 'AM'
pop_df.loc[pop_df['continent_code'] == 'SA', 'continent_code'] = 'AM'
pop_df.loc[pop_df['continent_code'] == 'OC', 'continent_code'] = 'AS'

pop_df.loc[pop_df['continent_code'] == 'AS', 'continent_code'] = 'Asia'
pop_df.loc[pop_df['continent_code'] == 'EU', 'continent_code'] = 'Europe'
pop_df.loc[pop_df['continent_code'] == 'AF', 'continent_code'] = 'Africa'
pop_df.loc[pop_df['continent_code'] == 'AM', 'continent_code'] = 'Americas'
````
